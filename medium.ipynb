```python
#q18

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def display(self):
        current = self.head
        elements = []
        while current:
            elements.append(str(current.data))
            current = current.next
        print(" -> ".join(elements) + " -> None")

    def insert_at_beginning(self, new_data):
        """Inserts a new node at the head of the list."""
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node
        print(f"Inserted {new_data} at the beginning.")

    def insert_at_end(self, new_data):
        """Inserts a new node at the end of the list."""
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            return

        last = self.head
        while last.next:
            last = last.next
        last.next = new_node
        print(f"Inserted {new_data} at the end.")

    def insert_after_node(self, prev_node_data, new_data):
        """Inserts a new node after a node with specific data."""
        current = self.head
        while current:
            if current.data == prev_node_data:
                new_node = Node(new_data)
                new_node.next = current.next
                current.next = new_node
                print(f"Inserted {new_data} after {prev_node_data}.")
                return
            current = current.next
        print(f"Node with data {prev_node_data} not found.")

llist_ins = LinkedList()
llist_ins.insert_at_end(10)
llist_ins.insert_at_beginning(5)
llist_ins.insert_at_end(30)
llist_ins.insert_after_node(10, 20)
llist_ins.display()
```

```python
#q19

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def display(self):
        current = self.head
        elements = []
        while current:
            elements.append(str(current.data))
            current = current.next
        print(" -> ".join(elements) + " -> None")

    def insert_at_end(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node
    
    def delete_node_by_value(self, key):
        """Deletes the first node with the given key (value)."""
        current = self.head
        
        if current and current.data == key:
            self.head = current.next
            print(f"Deleted head node with value {key}.")
            return

        prev = None
        while current and current.data != key:
            prev = current
            current = current.next

        if current is None:
            print(f"Value {key} not found in the list.")
            return

        prev.next = current.next
        print(f"Deleted node with value {key}.")

llist_del = LinkedList()
for val in [10, 20, 30, 40]:
    llist_del.insert_at_end(val)
print("Initial list:")
llist_del.display()

llist_del.delete_node_by_value(30)
llist_del.delete_node_by_value(10) 
llist_del.delete_node_by_value(50) 
print("List after deletions:")
llist_del.display()
```

```python
#q20

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    def count_nodes(self):
        """Counts the number of nodes in the linked list."""
        count = 0
        current = self.head
        while current:
            count += 1
            current = current.next
        return count

llist_count = LinkedList()
llist_count.insert_at_end(5)
llist_count.insert_at_end(15)
llist_count.insert_at_end(25)

print(f"The number of nodes in the list is: {llist_count.count_nodes()}")
```

```python
#q21

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    def search_element(self, key):
        """Searches for a key in the linked list and returns True/False."""
        current = self.head
        position = 0
        while current:
            if current.data == key:
                print(f"Element {key} found at position {position}.")
                return True
            current = current.next
            position += 1
        
        print(f"Element {key} not found in the list.")
        return False

llist_search = LinkedList()
for val in [10, 20, 30, 40]:
    llist_search.insert_at_end(val)

llist_search.search_element(30)
llist_search.search_element(50)
```

```python
#q23

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Stack:
    """Stack: Push/Pop from the head (O(1) complexity)."""
    def __init__(self):
        self.head = None

    def push(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
        print(f"Pushed: {data}")

    def pop(self):
        if self.head is None:
            return "Stack is empty"
        popped_data = self.head.data
        self.head = self.head.next
        return popped_data

    def peek(self):
        if self.head is None:
            return "Stack is empty"
        return self.head.data

print("--- Stack Implementation (LIFO) using Linked List ---")
my_stack = Stack()
my_stack.push(100)
my_stack.push(200)
print("Top element (Peek):", my_stack.peek())
print("Popped element:", my_stack.pop())
print("Top element after pop:", my_stack.peek())


class Queue:
    """Queue: Enqueue at tail, Dequeue from head (O(1) complexity)."""
    def __init__(self):
        self.head = None 
        self.tail = None 

    def enqueue(self, data):
        new_node = Node(data)
        if self.tail is None: 
            self.head = self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        print(f"Enqueued: {data}")

    def dequeue(self):
        if self.head is None:
            self.tail = None 
            return "Queue is empty"
        
        dequeued_data = self.head.data
        self.head = self.head.next
        return dequeued_data

print("\n--- Queue Implementation (FIFO) using Linked List ---")
my_queue = Queue()
my_queue.enqueue(1)
my_queue.enqueue(2)
print("Dequeued element:", my_queue.dequeue())
my_queue.enqueue(3)
print("Dequeued element:", my_queue.dequeue())
print("Dequeued element:", my_queue.dequeue())
print("Dequeued element:", my_queue.dequeue())
```
