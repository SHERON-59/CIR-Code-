```python
#q19
mark=float(input("Enter your mark"))
if(mark<=100 and mark>=90):
    grade='S'
elif(mark<=89 and mark>=85):
    grade='A+'
elif(mark<=84 and mark>=80):
    grade='A'
elif(mark<=79 and mark>=70):
    grade='B+'
elif(mark<=74 and mark>=70):
    grade='B'
elif(mark<=69 and mark>=60):
    grade='C'
elif(mark<=59 and mark>=50):
    grade='D'
elif(mark<=49 and mark>=40):
    grade='P'
else:
    grade='F'
print('marks=',mark,'\tgrade=',grade)
```
